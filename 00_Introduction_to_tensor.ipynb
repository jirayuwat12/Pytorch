```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Introduction to tensor

```python
# create scalar
scalar = torch.tensor(7)
print(scalar)
print('scalar.ndim = ', scalar.ndim) 
# get the number
print('scalar.item() = ', scalar.item())
```

```python
# create vector
vector = torch.tensor(data = [1,2,3])
print(vector)
print('vector.ndim = ', vector.ndim)
print('vector.shape = ', vector.shape)
```

```python
# create matrix
MATRIX = torch.tensor(data = [
    [1,0,0],
    [0,1,0],
])
print(MATRIX)
print('matrix.ndim = ', MATRIX.ndim) 
print('matrix.shape = ', MATRIX.shape)
```

## Create a random tensor

```python
# create random tensor
TENSOR = torch.rand(3,4)
print(TENSOR)
print('TENSOR.ndim = ', TENSOR.ndim)
print('TENSOR.shape = ', TENSOR.shape)
```

```python
# Create a random image
IMAGE = torch.rand(255,255,3)
plt.imshow(IMAGE)
```

## Manupulating tensor

```python
# Create tensor
a = torch.tensor([1,2,3])
b = torch.tensor([3,2,1])
c = torch.tensor([[1],[2],[3]])
```

```python
# adding
print(a+b)
print(a+10)
# substract
print(a-b)
print(a-10)
# multiplication
print(a*b)
print(a*10)
```

```python
# matrix multiplication
print(torch.matmul(a,c))
print(torch.matmul(a,b)) # why?  
```
